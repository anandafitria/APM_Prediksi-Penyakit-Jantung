# 3. Explanatory & Response Variables (Feature & Target)

```python
#konversi data ke numpy array 
#memasukkan data ke feature variable ke variable x
#dan target variable ke variable y

import numpy as np 

x=np.array(data[['Age','Sex','ChestPainType', 'RestingBP', 'Cholesterol','FastingBS', 'RestingECG', 'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope']])
y=np.array(data['HeartDisease'])
```

```python
#tampilkan data feature variable x

x
```

```python
#tampilkan data feature variable y

y
```

##### Preprocessing Dataset

```python
#labelencoder

from sklearn.preprocessing import LabelEncoder
import numpy as np 
le = LabelEncoder()
yle= le.fit_transform(y)
```

```python
#menampilkan dimensi x

x.shape
```

```python
#menampilkan dimensi y

yle.shape
```

# 4. Training & Testing Split Dataset

```python
#import library

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,yle , test_size=20, random_state=42)
```

```python
x_train
```

# 5. Training Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train, y_train)
```

# 6. Evaluasi Model Machine Learning

```python
pred=knn.predict(x_test)

#cetak hasil prediksi
pred
```

##### Accuracy Score

```python
from sklearn.metrics import accuracy_score 
final_model = KNeighborsClassifier(n_neighbors=3)
final_model.fit(x,yle)
print('Final Model Accuracy : %s' %accuracy_score(y_test, final_model.predict(x_test)))
```

```python
from sklearn.metrics import accuracy_score 
predic = final_model.predict(x_test)
accuracy_score(y_test, predic)
```

##### Classification Report

```python
from sklearn.metrics import classification_report, confusion_matrix
predic = final_model.predict(x_test)
print(classification_report(y_test, predic))
```

# 7. Deploy Model Machine Learning

```python
#KNN TANPA TUNING

knn_clf = KNeighborsClassifier(n_neighbors=2)
knn_clf.fit(x_train,y_train)

y_pred = knn_clf.predict(x_test)

round(accuracy_score(y_test,y_pred),2)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': np.arange(1,201)}

knn_clf = GridSearchCV(KNeighborsClassifier(), param_grid, cv=2, scoring='accuracy')

knn_clf.fit(x_train, y_train)
knn_clf.best_params_
```

```python
#KNN DENGAN TUNING TETANGGA TERDEKAT 33

knn_clf = KNeighborsClassifier(n_neighbors=33)
knn_clf.fit(x_train,y_train)

y_pred = knn_clf.predict(x_test)

round(accuracy_score(y_test,y_pred),2)
```

```python
#PICKLE

import pickle

with open('knn_pickle','wb') as r:
    pickle.dump(knn_clf,r)
```

```python
with open('knn_pickle','rb') as r:
    knnp = pickle.load(r)
```

```python
y_pred = knnp.predict(x_test)

round(accuracy_score(y_test,y_pred),2)
```
